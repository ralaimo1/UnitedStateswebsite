```python
import pandas as pd
import numpy as np
import matplotlib as sns
import matplotlib.pyplot as plt
from sklearn.linear_model import LinearRegression
import statsmodels.api as sm
```

```python
df = pd.read_csv('usa_data.csv')
```

```python
## I want to see a bar chart visualizing the number of orders placed for each Ua_browser
# I want to group all the instagram orders placed into a single data point for example (instagram has this mmany orders placed)

orders_placed_grouped = df.groupby('ua_browser')['total_orders_placed'].sum().reset_index()
orders_placed_grouped.plot.barh(x='ua_browser', y='total_orders_placed', color=['red','blue','green','yellow','orange'], legend=False)

plt.title("Total Orders Placed per Web Browser")
plt.xlabel("Number of orders placed")
plt.ylabel("Web browser category")

for index, value in enumerate(orders_placed_grouped['total_orders_placed']):
    plt.text(value, index, ' ' + str(value))
```
